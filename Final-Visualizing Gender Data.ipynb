```python
#Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading  Florida Data csv file
file_to_load = 'USSexDiffWages.csv'
#Converting to Pandas df
sex_data = pd.read_csv(file_to_load)
sns.set()
sex_data.head()
```

```python
#Plot data
plt.figure(figsize=(12,8))
plt.scatter(sex_data['female_%'], sex_data['female_wage'], alpha=0.9, s = sex_data['female_count']/80, color = "orange", marker = 6)
plt.scatter(sex_data['male_%'], sex_data['male_wage'], alpha=0.9, s = sex_data['male_count']/80, color = "royalblue", marker = 's')

#Labeling
plt.title('Gender Dichotomy in Post-Grad Labor Force')
plt.ylabel('Average Yearly Salary')
plt.xlabel('% in Major')

plt.title("Gender Dichotomy in Post-Grad Labor Force (2013-2017)", pad=10, fontsize=20)
plt.ylabel('Average Salary', fontsize = 14)
plt.xlabel('% in Major', fontsize = 14)
#Reference Lines
f_wage_mean = 56174.60
m_wage_mean = 37255.36
plt.axhline(y=m_wage_mean, color = 'red',alpha=.5)
plt.axhline(y=f_wage_mean, color = 'red', alpha=.5)
plt.legend(loc = 'best')
#plt.savefig("Figures/gen1")
```

```python
sex_data = sex_data.sort_values("female_wage", ascending=True)
plt.figure(figsize=(10,8))
plt.barh(sex_data["degree_name"].head(10), sex_data["female_wage"].head(10), color = "orange", height=.65)
plt.barh(sex_data["degree_name"].tail(10), sex_data["female_wage"].tail(10), color = "orange", height=.65)
plt.ylabel("Degree", fontsize=14)
plt.xlabel("Average Salary ($)", fontsize=14)
plt.title("Female Average Salary by Major: Highest / Lowest (2013-2017)", pad=10, fontsize=20)
plt.grid(axis="y")
plt.yticks(fontsize =15)
plt.axvline(x=f_wage_mean, color = "black", alpha=.5)
plt.axvline(x=m_wage_mean, color = "black", alpha=.5)
plt.text(30000,-1.5, "$%.0f" % f_wage_mean, fontsize = 12)
plt.text(48000,-1.5, "$%.0f" % m_wage_mean, fontsize = 12)
#plt.savefig("Figures/gen2")
plt.show()
```

```python
sex_data = sex_data.sort_values("male_wage", ascending=True)
plt.figure(figsize=(10,8))
plt.barh(sex_data["degree_name"].head(10), sex_data["male_wage"].head(10), color = "royalblue", height=.65)
plt.barh(sex_data["degree_name"].tail(10), sex_data["male_wage"].tail(10), color = "royalblue", height=.65)
plt.ylabel("Degree", fontsize=14)
plt.xlabel("Average Salary ($)", fontsize=14)
plt.title("Male Average Salary by Major: Highest / Lowest (2013-2017)", pad=10, fontsize=20)
plt.grid(axis="y")
plt.yticks(fontsize =15)
plt.axvline(x=m_wage_mean, color = "black", alpha=.5)
plt.axvline(x=f_wage_mean, color = "black", alpha=.5)
#plt.xticks(rotation=90, position= (10, 0), linespacing=1)
plt.text(48000,-1.5, "$%.0f" % m_wage_mean, fontsize = 12)
plt.text(30000,-1.5, "$%.0f" % f_wage_mean, fontsize = 12)
#plt.savefig("Figures/gen3")
plt.show()
```

```python
sex_data = sex_data.sort_values("male_%", ascending=True)
plt.figure(figsize=(10,8))
plt.barh(sex_data["degree_name"].head(10), sex_data["male_%"].head(10), color = "royalblue", height=.65)
plt.barh(sex_data["degree_name"].head(10), sex_data["female_%"].head(10), color = "orange", height=.65, left= sex_data['male_%'].head(10))
plt.barh(sex_data["degree_name"].tail(10), sex_data["male_%"].tail(10), color = "royalblue", height=.65)
plt.barh(sex_data["degree_name"].tail(10), sex_data["female_%"].tail(10), color = "orange", height=.65, left= sex_data['male_%'].tail(10))
plt.ylabel("Degree", fontsize=14)
plt.xlabel("% Male", fontsize=14)
plt.title("% Male: Highest / Lowest (2013-2017)", pad=10, fontsize=20)
plt.grid(axis="y")
plt.yticks(fontsize =15)
#plt.savefig("Figures/gen4")
plt.show()
```

```python
sex_data = sex_data.sort_values("female_count", ascending=True)
plt.figure(figsize=(10,5))
plt.barh(sex_data["degree_name"].tail(10), sex_data["female_count"].tail(10), color = "orange", height=.65)
plt.ylabel("Degree", fontsize=14)
plt.xlabel("Number of Female Graduates", fontsize=14)
plt.title("Female Populated Degrees: Highest (2013-2017)", pad=10, fontsize=20)
plt.grid(axis="x")
plt.yticks(fontsize =15)
plt.xticks(rotation=90, position= (10, 0), linespacing=1)
plt.show()
#plt.savefig("Figures/gen5")
```

```python
sex_data = sex_data.sort_values("male_count", ascending=True)
plt.figure(figsize=(10,5))
plt.barh(sex_data["degree_name"].tail(10), sex_data["male_count"].tail(10), color = "royalblue", height=.65)
plt.ylabel("Degree", fontsize=15)
plt.xlabel("Number of Male Graduates")
plt.title("Male Populated Degrees: Highest (2013-2017)", pad=10, fontsize=20)
plt.grid(axis="x")
plt.yticks(fontsize =15)
plt.xticks(rotation=90, position= (10, 0), linespacing=1)
plt.show()
#plt.savefig("Figures/gen6")
```
