```python
#This is a simple analysis of gender differences in post-secondary education
# * denotes incomplete task
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

college_majors = pd.read_csv("majors-list.csv")

recent_grads = pd.read_csv("recent-grads.csv")


#Combine the data into a single dataset
merged_df = pd.merge(college_majors, recent_grads, how = "left", on = ["Major"])
merged_df
```

```python
#Merge into dataframe, clean-up, remove columns
merged_df[['Major', 'Major_Category','Men', 'Women', 'Total']]
```

```python
#Group by major_category
major_groups = merged_df.groupby("Major_Category").sum()
major_groups
```

```python
#Consolidate into dataframe sorted by Major_category and gender
gender = major_groups[['Men', 'Women']]
gender
```

```python
#Plot with matplotlib
#*Fix Title - "Gender Dichotomy by Major Category"
#*Change orientation to horizontal stacked bar chart for easier reading
#*Change colors, add keys

gender.plot(kind = 'bar', stacked = True)
```
