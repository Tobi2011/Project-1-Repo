```python
df1_US_Edu_Wages = df1_US_Edu_Wages.sort_values("wage/salary", ascending=False)
SalaryMean = 47016.45
plt.figure(figsize=(10,5))
plt.barh(df1_US_Edu_Wages["degree_name"].head(10), df1_US_Edu_Wages["wage/salary"].head(10), color = "gold", height=.65)
plt.ylabel("Occupations", fontsize=10)
plt.xlabel("Average Salary ($)")
plt.title("Average Salary by Major: Highest (2013-2017)", pad=10, fontsize=20)
plt.grid(axis="x", alpha = .5)
plt.yticks(fontsize =12)
plt.axvline(x=SalaryMean, color = "red", alpha=.5)
#plt.xticks(rotation=90, position= (10, 0), linespacing=1)
plt.text(48000,6.8, "$%.0f" % SalaryMean, fontsize = 12)
plt.show()
```

```python
df1_US_Edu_Wages = df1_US_Edu_Wages.sort_values("wage/salary", ascending=True)
plt.figure(figsize=(10,5))
plt.barh(df1_US_Edu_Wages["degree_name"].head(10), df1_US_Edu_Wages["wage/salary"].head(10), color = "gold", height=.65)
plt.ylabel("Occupations", fontsize=10)
plt.xlabel("Average Salary ($)")
plt.title("Average Salary by Major: Lowest (2013-2017)", pad=10, fontsize=20)
plt.grid(axis="x")
plt.yticks(fontsize =12)
plt.axvline(x=SalaryMean, color = "red", alpha=.5)
plt.text(42500,7, "$%.0f" % SalaryMean, fontsize= 12)
plt.show()
```

```python
df1_US_Edu_Wages = df1_US_Edu_Wages.sort_values("NDJ_%", ascending=False)
NDJMean = 3.20
plt.figure(figsize=(10,5))
plt.barh(df1_US_Edu_Wages["degree_name"].head(10), df1_US_Edu_Wages["NDJ_%"].head(10), color = "royalblue", height=.65, alpha=.6)
#plt.ylabel("Top 10 Average Salary ($)")
plt.ylabel("Occupations", fontsize=10)
plt.xlabel("Non-Degree Jobs (%)")
plt.title("Grads /w Non-Degree Jobs: Highest (2013-2017)", pad=10, fontsize=20)
plt.yticks(fontsize =12)
plt.grid(axis="x")
#plt.xticks(rotation=90, position= (10, 0), linespacing=1)
plt.axvline(x=NDJMean, color = "red")
plt.text(NDJMean+ .2,6.8, "%.1f" % NDJMean + "%" , fontsize=12, backgroundcolor="white")
plt.show()
```
