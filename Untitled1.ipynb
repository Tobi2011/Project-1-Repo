```python
df1_US_Edu_Wages = df1_US_Edu_Wages.sort_values("wage/salary", ascending=False)

plt.barh(df1_US_Edu_Wages["degree_name"].head(10), df1_US_Edu_Wages["wage/salary"].head(10), color = "deepskyblue")
#plt.ylabel("Top 10 Average Salary ($)")
plt.title("Average Salary by Major: Highest (2013-2017)", pad=10, fontsize=13)
plt.grid(axis="x")
#plt.xticks(rotation=90, position= (10, 0), linespacing=1)
plt.show()
```

```python
df1_US_Edu_Wages = df1_US_Edu_Wages.sort_values("wage/salary", ascending=True)
#edgecolor = "black", linewidth=0.5
plt.figure(figsize=(10,5))
plt.barh(df1_US_Edu_Wages["degree_name"].head(10), df1_US_Edu_Wages["wage/salary"].head(10), color = "darkcyan")
plt.ylabel("Worst 10 Occupations")
plt.xlabel("Average Salary ($)")
plt.title("Average Salary by Major: Lowest (2013-2017)", pad=10, fontsize=20)
plt.grid(axis="x")
#plt.xticks(rotation=90, position= (10, 0), linespacing=1)
plt.show()
```
